# Comparison of HGC2 Chronologies

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('HGC2_Chronology_Compare.csv')
```

```python
data.head()
```

```python
# Define plot parameters
_ = plt.plot(data['Depth'], data['v2_TCE'], data['Depth'], data['v3_TCE'])

# Annotate the plot
_ = plt.title('Comparison of HGC2 Chronologies')
_ = plt.legend(('14C only', 'With Hurricanes'), loc='lower right')
_ = plt.xlabel('HGC2 Depth (cm)')
_ = plt.ylabel('Total Chronological Error (y)')

# Show the plot
#plt.show()

plt.savefig('HGC2_ChronologyCompare.png')
```

```python
# Define plot parameters
_ = plt.plot(data['Depth'], data['v3_Change'])

# Annotate the plot
_ = plt.title('Change to chronology with Hurricane addition')
_ = plt.xlabel('HGC2 Depth (cm)')
_ = plt.ylabel('Change in age (y)')

# Show the plot
#plt.show()

plt.savefig('HGC2_ChangeInAge.png')
```
